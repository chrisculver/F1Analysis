```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
import os

f1DataCache=os.path.join(os.path.expanduser('~'),'Documents','F1DataCache')
ff1.Cache.enable_cache(f1DataCache)

plotting.setup_mpl()
```

```python
#fuzzy matches race name, 'america also works'
weekend=ff1.get_event(2021, 'Imola') # type is weekend
print('Got data for {} on {}'.format(weekend['EventName'],weekend['EventDate']))
p1=weekend.get_practice(1) # type of all prac, q and r is Session
p2=weekend.get_practice(2)
p3=weekend.get_practice(3)
qual=weekend.get_qualifying()
race=weekend.get_race()
```

```python
laps=race.load()#Laps for all drivers
```

```python
nor=race.laps.pick_driver('NOR')
```

```python
type(nor)
```

```python
for lap in nor.iterlaps():
    fastf1Lap=lap[1]
    print(fastf1Lap.axes)
    break

compoundsUsed = []
lastLap=0
for lap in nor.iterlaps():
    if lap[1]['Compound'] not in compoundsUsed:
        compoundsUsed.append(lap[1]['Compound'])
    lastLap=lap[0]
        
print(compoundsUsed)
```

```python
dryTimes = []
telemetry = None

for lap in nor.iterlaps():
    fLap = lap[1]
    if fLap['Compound'] not in ['INTERMEDIATE']:
        if fLap['LapTime'] < ff1.utils.to_timedelta('85.0'):
            dryTimes.append(fLap['LapTime'])
            
    if lap[0]==lastLap:
        telemetry=lap[1].get_telemetry()
```

```python
import matplotlib.pyplot as plt

plt.plot(dryTimes)
plt.show()
```
