```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
import os

plotting.setup_mpl()

f1DataCache=os.path.join(os.path.expanduser('~'),'Downloads','F1DataCache')
ff1.Cache.enable_cache(f1DataCache)
```

```python
weekend=ff1.get_event(2022, 'bahrain')
race=weekend.get_race()
race.load()
```

```python
race.load()
```

```python
verLaps=race.laps.pick_driver('VER')
lecLaps=race.laps.pick_driver('LEC')
```

```python
fig, ax = plt.subplots()
ax.plot(verLaps['LapNumber'], verLaps['LapTime'], color='blue')
ax.plot(lecLaps['LapNumber'], lecLaps['LapTime'], color='red')
ax.set_title("VER vs LEC")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(verLaps['LapNumber'], verLaps['LapTime'], color=plotting.team_color('RedBull'))
ax.plot(lecLaps['LapNumber'], lecLaps['LapTime'], color=plotting.team_color('Ferrari'))
ax.set_title("VER vs LEC")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.set_ylim(ff1.utils.to_timedelta('94.0'),ff1.utils.to_timedelta('120.1'))
plt.show()
```
